```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Combine features (only 10 inputs, exclude 'cups' if 11 exist)
X_combined = pd.concat([X1, X2, X3, X4, X5, X6, X7, X8, X9, X10], axis=1)
Y1 = Y1.squeeze()  # make sure target is 1D (Series, not DataFrame)

# --- Feature scaling (very important for correct coefficient magnitudes)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_combined)

# --- Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, Y1, test_size=0.2, random_state=42
)

# --- Train model
model = LinearRegression()
model.fit(X_train, y_train)

# --- Model performance
print("Training R²:", model.score(X_train, y_train))
print("Test R²:", model.score(X_test, y_test))
print("\nIntercept (W0):", model.intercept_)

# --- Display coefficients
coef_df = pd.DataFrame({
    'Feature': X_combined.columns,
    'Coefficient (Wi)': model.coef_
})
print("\nCoefficients:\n", coef_df)

# --- Predict for new chocolate bread sample
new_cereal = [[49, 3.5, 0, 138, 17, 10, 0.5, 350, 10, 1]]  # 10 features
new_scaled = scaler.transform(new_cereal)
predicted_rating = model.predict(new_scaled)[0]

print("\nPredicted Rating for Chocolate Bread:", predicted_rating)

# --- Rating category
if predicted_rating < 30:
    print("Category: Poor cereal")
elif predicted_rating < 50:
    print("Category: Average cereal")
elif predicted_rating < 70:
    print("Category: Healthy cereal")
else:
    print("Category: Excellent cereal")

# --- Visualization (optional): Each feature vs Rating + Regression line
fig, axs = plt.subplots(3, 4, figsize=(20, 15))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

titles = ['Calories', 'Protein', 'Fat', 'Sodium', 'Fiber', 'Carbo', 
          'Sugars', 'Potass', 'Vitamins', 'Weight']

for idx, title in enumerate(titles):
    row = idx // 4
    col = idx % 4
    ax = axs[row, col]
    
    # inverse-transform feature for visualization
    x_feature = X_combined.iloc[:, idx]
    y_actual = Y1
    
    ax.scatter(x_feature, y_actual, alpha=0.5)
    ax.set_xlabel(title)
    ax.set_ylabel('Rating')
    ax.set_title(f'{title} vs Rating')

    # Regression line (approximation for visualization)
    x_sorted = np.linspace(x_feature.min(), x_feature.max(), 100).reshape(-1, 1)
    x_scaled = scaler.transform(np.hstack([
        np.zeros((100, X_combined.shape[1])),
    ]))  # baseline scaling
    x_scaled[:, idx] = scaler.transform(
        np.hstack([np.zeros((100, idx)), x_sorted, np.zeros((100, X_combined.shape[1]-idx-1))])
    )[:, idx]
    y_line = model.intercept_ + model.coef_[idx] * (x_scaled[:, idx])
    ax.plot(x_sorted, y_line, color='red', linewidth=2)

# Remove last empty subplot
axs[2, 3].remove()
plt.show()
```
