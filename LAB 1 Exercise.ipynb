```python
X = data.drop(columns=['id','stroke'])
X
```

```python
X = X.drop(columns=["Unnamed: 0"], errors="ignore")
X
```

applying dummy variable encoding

```python
X['gender']=X['gender'].replace('Male',1)
X['gender']=X['gender'].replace('Female',0)
X
```

```python
X['ever_married']=X['ever_married'].replace('Yes',1)
X['ever_married']=X['ever_married'].replace('No',0)
X
```

```python
X['Residence_type']=X['Residence_type'].replace('Urban',1)
X['Residence_type']=X['Residence_type'].replace('Rural',0)
X
```

```python
X['work_type'].unique()
```

applying ordinal encoding on work_type

```python
# One-Hot Encoding work_type
X = pd.get_dummies(X, columns=['work_type'], drop_first=True)
X
```

```python
X['smoking_status'].unique()
```

```python
# One-Hot Encoding work_type
X = pd.get_dummies(X, columns=['smoking_status'], drop_first=True)
X
```

```python
X["bmi"].fillna(X["bmi"].median(), inplace=True)
X
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

# pick numeric columns
num_cols = ["age", "avg_glucose_level", "bmi"]

scaler = StandardScaler()
df_scaled = X.copy()
df_scaled[num_cols] = scaler.fit_transform(X[num_cols])

print(df_scaled.head())
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))

for i, col in enumerate(num_cols):
    # original
    axes[0, i].hist(X[col], bins=30, color="skyblue", edgecolor="black")
    axes[0, i].set_title(f"Before Scaling: {col}")

    # scaled
    axes[1, i].hist(df_scaled[col], bins=30, color="salmon", edgecolor="black")
    axes[1, i].set_title(f"After Scaling: {col}")

plt.tight_layout()
plt.show()
```
